```python
#6.What are HTTP Methods used in Restful APIs?
#Ans:
#a)GET
#Purpose: Retrieve data from the server (read-only).
#Safe: Doesn’t change data on the server.

#b)POST
#Purpose: Create a new resource on the server.
#Not Idempotent: Multiple POSTs may create multiple resources.

#C)PUT
#Purpose: Update an entire resource (replace it with new data).
#Idempotent: Sending the same request multiple times results in the same state.
#d)PATCH
#Purpose: Partially update a resource (only specific fields).

#e)DELETE
#Purpose: Remove a resource from the server.
#Idempotent: Deleting the same resource multiple times has the same effect.
```

```python
#8. What is the difference between GET and POST HTTP method?
#Ans:
#a)Purpose
#GET → Retrieve data from the server (read-only).
#POST → Send data to the server to create or process something.

#b). Data Transmission
#GET → Sends data in the URL query string.
#POST → Sends data in the request body.
#c). Visibility
#GET → Data is visible in the URL (less secure).
#POST → Data is hidden in the body (more secure for sensitive info like passwords).

#c). Caching
#GET → Can be cached, bookmarked, and shared.
#POST → Not cached, not bookmarkable.

#d). Idempotency
#GET → Idempotent (calling it multiple times doesn’t change the resource).
#POST → Not idempotent (multiple POST requests can create multiple resources).

#e) Use Cases
#GET → Fetching data (e.g., search results, user profiles, product lists).
#POST → Submitting forms, creating new resources (e.g., new user, new order).
```

```python
#9.How do you handle errors in Flask API'S?
#Ans:Using abort()
from flask import Flask, abort, jsonify

app = Flask(__name__)

@app.route('/user/<int:user_id>')
def get_user(user_id):
    if user_id != 1:
        abort(404, description="User not found")
    return jsonify({"id": 1, "name": "Bhavya"})
```

```python
#b) Custom Error Handlers
@app.errorhandler(404)
def not_found(error):
    return jsonify({"error": "Resource not found"}), 404

@app.errorhandler(500)
def internal_error(error):
    return jsonify({"error": "Something went wrong on the server"}), 500
```

```python
#c)Try–Except Blocks
#d)Validation Errors
from flask import request

@app.route('/add_user', methods=['POST'])
def add_user():
    data = request.get_json()
    if not data or 'name' not in data:
        return jsonify({"error": "Name is required"}), 400
    return jsonify({"message": "User created", "data": data}), 201
```

```python
#e)Global Error Handling
@app.errorhandler(Exception)
def handle_exception(e):
    return jsonify({
        "error": str(e),
        "type": e.__class__.__name__
    }), 500
```

```python
#19.What are HTTP status codes,and why they are important in a flask?
#Ans:HTTP status codes are 3-digit numbers returned by a web server in response to a client’s request (like a browser, mobile app, or API client).
#They indicate whether the request was successful, failed, or needs additional action.
#Example:
#200 OK → Request succeeded
#404 Not Found → Requested resource does not exist
#500 Internal Server Error → Server crashed while processing
#In a Flask API or web app, status codes are crucial because:
#Communicate Result Clearly
#Tells the client (frontend, mobile app, API consumer) what happened with their request.
#Example: Returning 201 Created after creating a new user is more meaningful than just 200 OK.
#Error Handling

#Helps differentiate between client mistakes and server issues.
#Example: 400 Bad Request (client sent wrong data) vs 500 Internal Server Error (bug in server code).

#REST API Best Practices
#RESTful APIs rely heavily on proper status codes for clarity.
#Example: When deleting a resource, return 204 No Content instead of a generic success.

#Debugging & Logging
#Easier to monitor and debug issues by tracking codes like 404, 500, 401.

#Improves Client Behavior
#Some codes trigger specific actions in clients (e.g., 401 Unauthorized tells the client to re-authenticate).
```
