```python
#1.What is Simple Linear Regression(SLR)?Explain its purpose
```

```python
#Ans:Simple Linear Regression tries to find the best-fitting straight line (called the regression line) that predicts 
#the dependent variable (Y) from the independent variable (X).
#The general equation is:Y=b0​+b1​X+ε
#Where:
#Y = Dependent variable (what we want to predict)
#X = Independent variable (the predictor)
#b₀ = Intercept (value of Y when X = 0)
#b₁ = Slope (change in Y for a one-unit change in X)
#ε = Error term (difference between actual and predicted values)
#The main purposes of SLR are:
#Prediction
#To predict the value of one variable based on the value of another.
#Example: Predicting a person’s weight (Y) based on their height (X).
#Understanding Relationships
#To study how one variable affects or is associated with another.
#Example: Understanding how study hours influence exam scores.
#Trend Estimation
#To find and describe linear trends in data over time or across categories.
#Example: Estimating the growth trend in sales over months.
#Quantifying the Effect
#The slope (b₁) quantifies the impact of X on Y (e.g., “for every 1 unit increase in X, Y changes by b₁ units”).
```

```python
#2.What are the Key assumption of Simple Linear Regression?
```

```python
#Ans:Assumptions of Simple Linear Regression
#Linearity
#Assumption: The relationship between the independent variable (X) and the dependent variable (Y) is linear.
#Meaning: As X increases, Y changes at a constant rate.
#How to Check: Plot X vs Y — if the scatterplot shows a straight-line pattern, the assumption holds.
#If Violated: The model will not fit well — consider using polynomial or nonlinear regression.

# Independence of Errors
#Assumption: The residuals (errors) are independent of each other.
#Meaning: The error for one observation should not depend on another.
#Example: In time series data, consecutive observations may be correlated (autocorrelation).
#How to Check: Use Durbin–Watson test (value near 2 indicates independence).
#If Violated: Use time-series models (like ARIMA) or include lag variables.

# Homoscedasticity (Constant Variance of Errors)
#Assumption: The residuals have constant variance across all levels of X.
#Meaning: The spread of errors should be roughly the same for small and large X values.
#How to Check: Plot residuals vs predicted values — the spread should be uniform (no funnel shape).
#If Violated: It indicates heteroscedasticity; use weighted least squares or transform variables (e.g., log(Y)).

# Normality of Errors
#Assumption: The residuals (errors) are normally distributed.
#Meaning: The differences between actual and predicted Y values follow a normal distribution.
#How to Check: Use a Q–Q plot or histogram of residuals.
#If Violated: Confidence intervals and significance tests may become unreliable.
#Can fix using log/square root transformation or robust regression.

# No (or minimal) Measurement Error in X
#Assumption: The independent variable (X) is measured without error.
#Meaning: X values are assumed to be exact.
#If Violated: It can bias slope estimates — this issue is called “errors-in-variables” problem.
```

```python
#3.Write the methamatical equation for a simple linear regression model and explain the each term?
#Ans:Mathematical Equation of Simple Linear Regression
#Yi=β0+β1Xi+εi
#Yi: Dependent Variable:The value we want to predict or explain.
#Xi:Independent Variable:The predictor or input variable used to explain Y.
#β0:Intercept (Constant term):he predicted value of Y when X = 0.
#β1:Slope Coefficient:The change in Y for a one-unit change in X.
#εi:Error Term (Residual):The difference between the actual and predicted Y for observation i.
```

```python
#5. What is the method of least squares in linear regression?
#Ans:The Method of Least Squares finds the regression line that minimizes the total squared differences (errors)
#between the actual values of the dependent variable (Y) and the predicted values (Y cap)the line.
#In other words, it makes the sum of squared errors (SSE) as small as possible
```

```python
#6.What is logistic Regression?How does it differ from linear regression?
#Ans:Logistic Regression is a statistical and machine learning technique used for classification problems, especially when the dependent variable (Y)
#is categorical (usually binary).
#| **Aspect**                  | **Linear Regression**                     | **Logistic Regression**                                        |
# --------------------------- | ----------------------------------------- | -------------------------------------------------------------- |
# **Purpose**                 | Predicts a **continuous** numeric outcome | Predicts a **categorical (binary or multi-class)** outcome     |
# **Output Range**            | Any real number (−∞ to +∞)                | Probability between **0 and 1**                                |
# **Equation Type**           | Linear equation: (Y = b_0 + b_1X)         | Logistic (sigmoid) function: (P = \frac{1}{1+e^{-(b_0+b_1X)}}) |
# **Error Metric**            | Uses **Least Squares** (minimize SSE)     | Uses **Maximum Likelihood Estimation (MLE)**                   |
# **Line Type**               | Straight line                             | S-shaped (sigmoid) curve                                       |
# **Use Case Example**        | Predicting house prices, sales, weight    | Predicting disease presence, email spam, loan approval         |
# **Dependent Variable Type** | Continuous                                | Categorical (0/1, Yes/No, etc.)                                |
```

```python
#8.What is the purpose of the R squared metric in regression analysis?
#Ans:R-squared represents the proportion of variance in the dependent variable (Y) that is explained by the independent variable(s) (X) in the model.
#n simple terms — it tells us how much of the variation in Y can be predicted from X.
#Interpretation
#| **R² Value**    | **Meaning**                                            |
#| --------------- | ------------------------------------------------------ |
#| **0**           | Model explains **none** of the variance in Y           |
#| **1**           | Model explains **all** the variance in Y (perfect fit) |
#| **0.7**         | Model explains **70%** of the variation in Y           |
#| **Closer to 1** | Better the model’s explanatory power                   |
```
