```python
# Qno.1 What is Anomaly Detection? Explain its types with examples
#Ans:Anomaly Detection is a data analysis technique used to identify data points, events, or patterns that deviate significantly from normal behavior. These unusual patterns—called anomalies or outliers—often indicate critical events such as fraud, system failure,
#security breaches, or medical abnormalities.
#ypes of Anomaly Detection (with Examples)
#1. Point Anomalies
#A single data point is anomalous compared to the rest of the data.
# Example:
#In a bank transaction system, a customer usually withdraws ₹2,000–₹5,000, but suddenly withdraws ₹2,00,000.
#A patient’s normal heart rate is 70–80 bpm, but one reading shows 180 bpm.
#2. Contextual Anomalies (Conditional Anomalies)
#A data point is anomalous only in a specific context, such as time, location, or season.
# Example:
#30°C temperature is normal in summer but anomalous in winter.
#High electricity usage at 2 PM is normal for factories but abnormal at 2 AM.
#3. Collective Anomalies
#A group of data points together is anomalous, even if individual points appear normal.
# Example:
#A sequence of small network packets may indicate a DDoS attack.
#Multiple failed login attempts in a short time window.
```

```python
#QNO2: Compare Isolation Forest,DBSCAN and Local Outlier Factor in the terms of their approach and suitable use cases?
#1. Isolation Forest (IF)
#Approach
#Based on the idea that anomalies are easier to isolate than normal points.
#Uses random feature selection and random split values to build decision trees.
#Anomalies require fewer splits to be isolated → shorter path length.
#No distance or density calculation.
#Key Characteristics
#Works well in high-dimensional data
#Fast and scalable (O(n log n))
#Global anomaly detection
#Assumes anomalies are rare and different
#Suitable Use Cases
# Fraud detection in financial transactions
# Network intrusion detection
# Large healthcare datasets (lab results, claims data)
# Sensor anomaly detection in IoT

#2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
#Approach
#Density-based clustering algorithm.
#Forms clusters based on ε (eps) neighborhood and MinPts.
#Points not belonging to any cluster are labeled as noise (anomalies).
#Anomalies lie in low-density regions.
#Key Characteristics
#Does not require number of clusters
#Can detect arbitrarily shaped clusters
#Sensitive to parameter tuning (ε, MinPts)
#Struggles with varying density and high-dimensional data
#Suitable Use Cases
# Spatial data analysis
# GPS trajectory anomalies
# Image segmentation
# Network traffic bursts

#3. Local Outlier Factor (LOF)
#Approach
#Density-based but local.
#Compares density of a point to its k-nearest neighbors.
#If a point has much lower density than neighbors → anomaly.
#Captures local anomalies that global methods miss.
#Key Characteristics
#Excellent for local deviations
#Sensitive to k (number of neighbors)
#Computationally expensive for large datasets
#Poor performance in very high dimensions
#Suitable Use Cases
# Healthcare monitoring (patient vitals)
# Credit card fraud in customer-specific patterns
# Network behavior monitoring
# Small-to-medium datasets
    
```

```python
# Qno3: What are the key component of a Time series?
#Ans:A time series is a sequence of observations recorded at regular time intervals (hourly, daily, monthly, yearly, etc.).
#Its behavior is commonly explained using four key components.
#Key Components of a Time Series
#1. Trend (T)
#The long-term movement of the series over time.
#Shows whether the data is generally increasing, decreasing, or stable.

# Example:

#Gradual increase in hospital OPD registrations over several years
#Rising monthly electricity consumption

#2. Seasonality (S)

#Regular, repeating patterns at fixed intervals (daily, monthly, yearly).
#Caused by calendar effects, weather, habits, or policies.
# Example:
#Increase in dengue cases during monsoon
#Higher retail sales during festive seasons
#Higher patient inflow on weekdays than weekends

# Occurs at known and fixed periods.

#3. Cyclical Component (C)
#Long-term fluctuations around the trend.
#Do not have a fixed or known period.
#Usually linked to economic or social cycles.
# Example:
#Economic boom and recession cycles
#Long-term changes in health expenditure patterns

#4. Irregular / Random Component (R)
#Unpredictable, random variations.
#Caused by unexpected events like disasters, strikes, epidemics.

#Example:

#Sudden spike in hospital admissions during COVID-19
#Drop in traffic due to sudden lockdown
```
